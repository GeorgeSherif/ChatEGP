```python
import matplotlib.pyplot as plt
import numpy as np

# Data for the bar chart
classes = ['English Dataset (Training)', 'Arabic Dataset (Testing)']
subclasses = ['Before Preprocessing', 'After Preprocessing']
values = np.array([[86.7, 82], [87, 78]])

# Number of subclasses and their width
num_subclasses = len(subclasses)
width = 0.4

# Plotting the grouped bar chart
fig, ax = plt.subplots()
for i in range(num_subclasses):
    ax.bar(np.arange(len(classes)) + i * width, values[:, i], width=width, label=subclasses[i])

    # Adding labels on top of the bars
    for j, v in enumerate(values[:, i]):
        ax.text(j + i * width, v + 1, str(v), ha='center')

# Adding labels and title
#ax.set_xlabel('Classes')
#ax.set_ylabel('Values')
ax.set_title('Accuracies Comparison')
ax.set_xticks(np.arange(len(classes)) + (num_subclasses - 1) * width / 2)
ax.set_xticklabels(classes, ha='center')
ax.legend()
ax.set_ylim(bottom=75)
ax.set_yticks(np.arange(75, np.max(values) + 10, 10))
ax.grid(False)

# Displaying the plot
plt.show()
```

```python
import matplotlib.pyplot as plt

# Example sequence of numbers
Epochs = ["Epoch 1", "Epoch 2", "Epoch 3","Epoch 4", "Epoch 5","Epoch 6", "Epoch 7","Epoch 8"]
sequence = [67.7, 81.8, 85.1, 83.4, 85.1, 86.7, 86.2, 85.9]

fig, ax = plt.subplots(figsize=(9, 4))

# Plotting the sequence as a line plot
plt.plot(Epochs, sequence)
ax.grid(False)

# Adding numbers on the nodes
for i, value in enumerate(sequence):
    plt.text(i, value, str(value), ha='center', va='bottom')
# Adding labels and title
plt.xlabel('Index')
plt.ylabel('Value')
plt.title('Accuracies of Training Epochs Before Preprocessing (English Dataset)')

ax.set_ylim(bottom=65)
ax.set_ylim(top=90)

# Displaying the plot
plt.show()
```
