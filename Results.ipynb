```python
import matplotlib.pyplot as plt
import numpy as np

# Data for the bar chart
classes = ['English Dataset (Training)', 'Arabic Dataset (Testing)']
subclasses = ['Without Preprocessing', 'Using Preprocessing', 'Using Data Augmentation']
values = np.array([[87.1, 81.8, 87.6], 
                   [83.8, 81.3, 86.8]])

# Number of subclasses and their width
num_subclasses = len(subclasses)
width = 0.2

# Plotting the grouped bar chart
fig, ax = plt.subplots()
for i in range(num_subclasses):
    ax.bar(np.arange(len(classes)) + i * width, values[:, i], width=width, label=subclasses[i])

    # Adding labels on top of the bars
    for j, v in enumerate(values[:, i]):
        ax.text(j + i * width, v + 1, str(v), ha='center')

# Adding labels and title
#ax.set_xlabel('Classes')
#ax.set_ylabel('Values')
ax.set_title('Accuracies Comparison')
ax.set_xticks(np.arange(len(classes)) + (num_subclasses - 1) * width / 2)
ax.set_xticklabels(classes, ha='center')
ax.legend()
ax.set_ylim(bottom=75)
ax.set_yticks(np.arange(75, np.max(values) + 10, 10))
ax.grid(False)

# Displaying the plot
plt.show()
```
